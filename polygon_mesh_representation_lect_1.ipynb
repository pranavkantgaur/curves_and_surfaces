## Concepts:
  * Objects:
    * Polygonal(piecewise-linear): More efficient with scale. Will be covered in later lectures.
    * Polynomial(piecewise-smooth): More _realism_ at cost of the computation. Preferred for smaller geometries. **Focus of current lectures**
  * Mathematical representations:
    * Explicit, $y = f(x)$
    * Implicit, $f(x, y, z) = 0$
    * Parametric: Q(t) = [x(t) y(t) z(t)]
